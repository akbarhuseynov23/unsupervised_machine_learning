# Scaling with Scikit-Learn Transformers

## MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler(feature_range=(0,1))
minmax.fit(music)
music_scaled = minmax.transform(music)
```

```python
music_scaled = pd.DataFrame(music_scaled, columns = music.columns, index = music.index)
music_scaled
```

### Outliers

```python
plt.subplots(figsize=(16, 10))
sns.boxplot(data=music_scaled.loc[:, music.columns])
plt.title('MinMaxScaler')
plt.show()
```

## StandardScaler

```python
from sklearn.preprocessing import StandardScaler
std_scler = StandardScaler()
std_scler.fit(music_scaled)
std_scler.transform(music_scaled)
```

```python
#alternatively:
std_scaled_music = std_scler.fit_transform(music_scaled)
```

```python
std_scaled_music = pd.DataFrame(std_scaled_music, columns = music.columns, index = music.index)
std_scaled_music
```

### Outliers

```python
plt.subplots(figsize=(14, 8))
sns.boxplot(data=std_scaled_music.loc[:, music.columns])
plt.title('StandardScaler')
plt.show()
```

## RobustScaler()

This Scaler removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).

Centering and scaling happen independently on each feature by computing the relevant statistics on the samples in the training set. Median and interquartile range are then stored to be used on later data using the transform method.

```python
from sklearn.preprocessing import RobustScaler

transformer = RobustScaler()
music_scaled = transformer.fit_transform(music_scaled)

# transformer = RobustScaler().fit(music_scaled)
# transformer.transform(music_scaled)
```

```python
music_robus_scaled = pd.DataFrame(music_scaled, columns = music.columns, index = music.index)
music_robus_scaled
```

### Outliers

```python
plt.subplots(figsize=(14, 8))
sns.boxplot(data=music_robus_scaled.loc[:, music.columns])
plt.title('RobustScaler')
plt.show()
```

## QuantileTransformer()

This method transforms the features to follow a uniform or a normal distribution. Therefore, for a given feature, this transformation tends to spread out the most frequent values. It also reduces the impact of (marginal) outliers: this is therefore a robust preprocessing scheme.

The transformation is applied on each feature independently. First an estimate of the cumulative distribution function of a feature is used to map the original values to a uniform distribution. The obtained values are then mapped to the desired output distribution using the associated quantile function. 

```python
import numpy as np
from sklearn.preprocessing import QuantileTransformer

qtl_transformer = QuantileTransformer(n_quantiles=10, random_state=0)
music_qtl_scaled = qtl_transformer .fit_transform(music)
music_qtl_scaled = pd.DataFrame(music_qtl_scaled, columns = music.columns, index = music.index)
music_qtl_scaled
```

### Outliers

```python
plt.subplots(figsize=(12, 8))
sns.boxplot(data=music_qtl_scaled.loc[:, music.columns])
plt.title('QuantileTransformer')
plt.show()
```

## K-means

```python
# Import the model from the pertinent submodule, in this case sklearn.cluster.
from sklearn.cluster import KMeans
```

```python
# Initialize the model with the parameters
kmeans = KMeans(n_clusters=5)
```

```python
# Fit the model to the data using its fit method
kmeans.fit(music_scaled)
```

```python
# Obtain the cluster for each observation with the predict method
clusters = kmeans.predict(music_scaled)
```

```python
# Attach the cluster output to thr original DataFrame
music["cluster"] = clusters
music
```

```python
# Using groupbymethod from Pandas can quickly find out the average value that each cluster has for each one of the features in use
music.groupby(by="cluster").mean() 
```

```python
import pandas as pd
```

```python
# Access centroids of the cluster, which had already been computed in by kmeans model, and are stored as an attribute of the model.
centroids = kmeans.cluster_centers_
pd.DataFrame(centroids)
```

```python
plt.subplots(figsize =(8, 6))
sns.heatmap(centroids)
```
