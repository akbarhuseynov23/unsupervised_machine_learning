## Manhattan_distance

```python
music_distance = music.iloc[0:,1:13]
results = {
    'comparison': [],
    'Manhattan_distance': []
}
for i in range(len(music_distance.index)):
    for j in range(i, len(music_distance.index)):
        if i != j:
            results['comparison'].append(music_distance.iloc[i].name+' vs '+music_distance.iloc[j].name)
            results['Manhattan_distance'].append(sum(abs(music_distance.iloc[i] - music_distance.iloc[j])))
            
results = pd.DataFrame(results).set_index('comparison')
results
```

```python
all_distances = []

for song in music.index: 
    
    # get the rating for the first three columns 
    danceability = music.loc[song][1]
    energy = music.loc[song][2]
    key = music.loc[song][3]
    loudness = music.loc[song][4]
    mode = music.loc[song][5]
    speechiness = music.loc[song][6]
    acousticness = music.loc[song][7]
    instrumentalness = music.loc[song][8]
    liveness = music.loc[song][9]
    valence = music.loc[song][10]
    tempo = music.loc[song][11]
    time_signature = music.loc[song][12]
    

    print(f'{song} ratings for danceability: {danceability}, energy: {energy}, key: {key}, loudness: {loudness}, mode: {mode}, speechiness: {speechiness}, acousticness: {acousticness}, acousticness: {acousticness}, instrumentalness: {instrumentalness}, liveness: {liveness}, valence: {valence}, tempo: {tempo}, time_signature: {time_signature}')
    
    song_distance = []

    # now compare its ratings with everyone else: 
    for song_to_compare in music.index:
        
        # get the ratings for the other songs to compare
        danceability_to_compare = music.loc[song_to_compare][1]
        energy_to_compare = music.loc[song_to_compare][2]
        key_to_compare = music.loc[song_to_compare][3]
        loudness_to_compare = music.loc[song_to_compare][4]
        mode_to_compare = music.loc[song_to_compare][5]
        speechiness_to_compare = music.loc[song_to_compare][6]
        acousticness_to_compare = music.loc[song_to_compare][7]
        instrumentalness_to_compare = music.loc[song_to_compare][8]
        liveness_to_compare = music.loc[song_to_compare][9]
        valence_to_compare = music.loc[song_to_compare][10]
        tempo_to_compare = music.loc[song_to_compare][11]
        time_signature_to_compare = music.loc[song_to_compare][12]
        
        # look for the distances between both songs
        danceability_distance = abs(danceability - danceability_to_compare)
        energy_distance = abs(energy - energy_to_compare)
        key_distance = abs(key - key_to_compare)
        loudness_distance = abs(loudness - loudness_to_compare)
        mode_distance = abs(mode - mode_to_compare)
        speechiness_distance = abs(speechiness - speechiness_to_compare)
        acousticness_distance = abs(acousticness - acousticness_to_compare)
        instrumentalness_distance = abs(instrumentalness - instrumentalness_to_compare)
        liveness_distance = abs(liveness - liveness_to_compare)
        valence_distance = abs(valence - valence_to_compare)
        tempo_distance = abs(tempo - tempo_to_compare)
        time_signature_distance = abs(time_signature - time_signature_to_compare)
        
        # add the distances between both students
        distance = danceability_distance + energy_distance + key_distance + loudness_distance + mode_distance + speechiness_distance + acousticness_distance + instrumentalness_distance + liveness_distance + valence_distance + tempo_distance + time_signature_distance
        
        print(f'The distance between {song} and {song_to_compare} is {distance}')
        
        # store the distance
        song_distance.append(distance)
        
    # the list of distances of an specific student
    print(f'The distances from {song} are {song_distance}')
    
    # store the student distances in all the students distances
    all_distances.append(song_distance)
```

```python
# adding the column names to understand the table
print('Distances between songs ')
# Create a dataframe from the results
song_manhattan_distance = pd.DataFrame(all_distances)
# Label the columns using the names in our original dataframe (music)
song_manhattan_distance.columns = music.index
# Label the rows using the names in our original dataframe (music)
song_manhattan_distance.index = music.index

song_manhattan_distance
```

## Euclidean distance

```python
import math

sample = music.iloc[0:,1:13]
heatmap_results = {
    'song_1': [],
    'song_2': [],
    'Euclidean_distance': []
}
results = {
    'comparison': [],
    'Euclidean_distance': []
}

for i in range(len(sample.index)):
    for j in range(len(sample.index)):
        heatmap_results['song_1'].append(sample.iloc[i].name)
        heatmap_results['song_2'].append(sample.iloc[j].name)
        heatmap_results['Euclidean_distance'].append(math.sqrt(sum((sample.iloc[i] - sample.iloc[j])**2)))
        
        if i != j:
            results['comparison'].append(sample.iloc[i].name+' vs '+sample.iloc[j].name)
            results['Euclidean_distance'].append(math.sqrt(sum((sample.iloc[i] - sample.iloc[j])**2)))
            
song_Euclidean_distance = pd.DataFrame(heatmap_results)
song_Euclidean_distance_pivot_table = pd.pivot_table(song_Euclidean_distance, 
                                                        values='Euclidean_distance', 
                                                        index='song_1',
                                                        columns='song_2')
plt.subplots(figsize=(12,8))
sns.heatmap(song_Euclidean_distance_pivot_table)

results = pd.DataFrame(results).set_index('comparison')
results
```

#### Function to calculate Euclidean or Manhatten

```python
def distance_music(data, dist_type='euclidean'):

    from math import sqrt
    all_distances = []
    error = False
    
    for song in data.index: 

        ratings = [rating for rating in data.loc[song]]
        song_distances = []

        for song_to_compare in data.index: 
            ratings_to_compare = [rating_to_compare for rating_to_compare in data.loc[song_to_compare]]

            if dist_type == 'euclidean':
                distance = sqrt(sum([(ratings[pos] - ratings_to_compare[pos])**2 for pos in range(len(ratings))]))
            elif dist_type == 'manhattan':
                distance = sum([abs(ratings[pos] - ratings_to_compare[pos]) for pos in range(len(ratings))])
            else: 
                print('Wrong distance type. Please choose between "manhattan" or "euclidean" distance.')
                error = True
                break
                
            song_distances.append(distance)
            
        if error: break
        
        all_distances.append(song_distances)
        
    if error is False: 
        st_euclidean_distances = pd.DataFrame(all_distances)
        st_euclidean_distances.columns = music.index
        st_euclidean_distances.index = music.index
        return st_euclidean_distances
```

## Data Scaling

```python
music = pd.read_csv('../data/df_audio_features.csv')
music.set_index(['song_name', 'artist'], inplace=True)
music
```

```python
def min_max_scaling(col):
    min_col = min(col)
    max_col = max(col)
    norm_val = []
    for val in col:
        norm_val.append((val - min_col) / (max_col - min_col))
    return norm_val
```

```python
music_norm = music.apply(min_max_scaling, axis=0).copy()
music_norm
```

```python
sns.heatmap(distance_music(music))
plt.title('No scaling applied')
plt.show()
```

```python
sns.heatmap(distance_music(music_norm))
plt.title('Min & Max Scaled')
plt.show()
```
